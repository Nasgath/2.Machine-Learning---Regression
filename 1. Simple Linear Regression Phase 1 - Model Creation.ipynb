```python
# pandas is a library we are impporting library pandas and assigned as pd

#it supports tables,file read,functions and it is globalized

import pandas as pd
```

```python
#dataset is variable
#pandas is assigned as pd previously so we are using pd here 
#and it used to read the file name as 'Salary_Data.csv' with help of read function pd.read_csv.

dataset = pd.read_csv("Salary_Data.csv")
```

```python
#to print the values under dataset

dataset
```

```python
#independent is a varilable
#To seprate the particular column in the dataset table
#dataset[['YearsExperience']] - in dataset table 'YearsExperience' is alone is assigned to the  variable independent
independent = dataset[['YearsExperience']]

#to print indepentent variable
independent

```

```python
#dependent is a varilable
#To seprate the particular column in the dataset table
#dataset[['Salary']] - in dataset table 'SAlary' is alone is assigned to the  variable depentent

dependent = dataset[['Salary']]
#to print dependent variable
dependent
```

```python
# for ML sklearn is a provider

#  to split our data into train and test sets , X_train,X_test = input  and Y_trian,Y_test = output

from sklearn.model_selection import train_test_split

X_train,X_test,Y_trian,Y_test = train_test_split(independent,dependent, test_size = 0.30, random_state = 0)
```

```python
# it a learning phase 

#regressor ia a variable
#First 2 line is a procedure, X_train,Y_trian = substitue place

from sklearn.linear_model import LinearRegression

regressor =  LinearRegression() # LinearRegression() is a class and is assigned to variable regressor

regressor.fit (X_train,Y_trian) # inside the regressor fit function is called, X_train,Y_trian = training set input and output
```

```python
#weight is a variable, in coeffiencent method weight is calculated and assigned to variable weight.
weight = regressor.coef_
weight #to print weight varibal value
```

```python
# bias is a variable, in intercept method bias is calculated and assigned to variable bias.
bias = regressor.intercept_
bias #to print bias varibal value 

#weight, bias,model are saved under regressor variable
```

```python
#it a crosscheck process our brain predicted,X_test is a test data and it is not avaialble in training data
#Y_pred is a variable used for predecion of out test set and in regressor variable we have everything and
#.predict is a function is used
Y_pred = regressor.predict(X_test)

```

```python
#r2_score  is function
#Y_test = actual output ,Y_pred = predicted output

from sklearn.metrics import r2_score

r_score = r2_score (Y_test,Y_pred)
```

```python
 #to check the value of r_score 

#if r_score is nearly to 1 good model and r_score is nearly to 0 bad model

r_score
```

```python
#to save a model we use pickle function and pickle extension is .sav

import pickle

filename = 'Finalized_Model_Linear.sav' #dummy file name is created as filename
```

```python
#we are dumpping the model regressor in pickle

pickle.dump(regressor,open(filename,'wb')) #wb - write binary

#file will be created in the name of 'Finalized_Model_Linear.sav' under the folder.
```

```python
#we are going to load the saved model to send to phase 2 
#loaded_model is a variable name and pickle.load is loaded in pickle with file name Finalized_Model_Linear.sav in read mode.
#in result variable we are saving the loaded model and 
#we are giving the values as years of experience in predict and it show the salary in the result
loaded_model = pickle.load(open("Finalized_Model_Linear.sav",'rb'))
result = loaded_model.predict([[13]])
```
